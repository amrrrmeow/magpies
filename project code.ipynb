```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из файлов
df_wing_length = pd.read_excel('C:/Users/vn_al/OneDrive/Рабочий стол/длина крыла взрослые самки.xlsx')
df_tail_length = pd.read_excel('C:/Users/vn_al/OneDrive/Рабочий стол/длина хвоста взрослые самки.xlsx')

# Построение диаграммы
fig, ax = plt.subplots()
ax.bar(df_wing_length.columns, df_wing_length.iloc[0])
ax.set_ylabel('Длина крыла (мм)')
ax.set_title('Длина крыла взрослые самки')

fig, ax = plt.subplots()
ax.bar(df_tail_length.columns, df_tail_length.iloc[0])
ax.set_ylabel('Длина хвоста (мм)')
ax.set_title('Длина хвоста взрослые самки')

plt.show()
```
